```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load the CSV file
csv_path = '/Users/donzano/Downloads/EPA_SmartLocationDatabase_V3_Jan_2021_Final.csv'  # Update this path

# Read CSV into DataFrame
df = pd.read_csv(csv_path)

# Step 2: Explore the columns
print("Available columns:")
print(df.columns.tolist())

# Step 3: Define numeric columns to plot
numeric_columns = ['TotPop', 'AutoOwn0', 'Pct_AO0', 'R_PCTLOWWAGE', 'NatWalkInd']

# Step 4: Plot histograms for selected columns
for column in numeric_columns:
    plt.figure(figsize=(10,6))
    df[column].hist(bins=30)
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.grid(False)
    plt.show()

# Step 5: Plot scatter plot comparing Pct_AO0 (households with zero cars) to R_PCTLOWWAGE (percentage low wage workers)
plt.figure(figsize=(10,6))
plt.scatter(df['Pct_AO0'], df['R_PCTLOWWAGE'], color='green', alpha=0.5)
plt.title('Percentage of Low Wage Workers vs Households with Zero Cars')
plt.xlabel('Percentage of Households with Zero Cars (Pct_AO0)')
plt.ylabel('Percentage of Low Wage Workers (R_PCTLOWWAGE)')
plt.grid(True)
plt.show()

# Step 6: Plot scatter plot comparing Total Population to Walkability Index
plt.figure(figsize=(10,6))
plt.scatter(df['TotPop'], df['NatWalkInd'], color='purple', alpha=0.5)
plt.title('Total Population vs Walkability Index')
plt.xlabel('Total Population')
plt.ylabel('Walkability Index (NatWalkInd)')
plt.grid(True)
plt.show()

# Step 7: Plot scatter plot comparing Total Population to Households with Zero Cars
plt.figure(figsize=(10,6))
plt.scatter(df['TotPop'], df['Pct_AO0'], color='orange', alpha=0.5)
plt.title('Total Population vs Percentage of Households with Zero Cars')
plt.xlabel('Total Population')
plt.ylabel('Percentage of Households with Zero Cars (Pct_AO0)')
plt.grid(True)
plt.show()
```
