```python
# Import libraries
import geopandas as gpd
import fiona

# Set the path to the .gdb directory
gdb_path = '/Users/donzano/Downloads/Natl_WI.gdb'  # Update this path

# Step 1: List all available layers in the GDB
layers = fiona.listlayers(gdb_path)
print("Available Layers:")
for layer in layers:
    print("-", layer)

# Step 2: Load a specific layer (replace with actual layer name after checking the list)
layer_name = layers[0]  # You can change the index or name based on the layer you're interested in
gdf = gpd.read_file(gdb_path, layer=layer_name)

# Step 3: Preview the data
print("\nData Preview:")
print(gdf.head())

# Step 4: Summary of the dataframe
print("\nDataFrame Info:")
print(gdf.info())

# Step 5: Plot if it has geometry
if 'geometry' in gdf.columns:
    gdf.plot()

# Optional: Save to CSV for non-spatial analysis
#gdf.to_csv('walkability_layer.csv', index=False)
```
